# Class 02, linearization
- find fixed points symbolically
- find fixed points numerically
- plot $f(x)$ from $\frac{dx}{dt} = f(x(t))$ in order to visualize $\frac{dx}{dt}$

```python
import sympy as sy
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt

# Find fixed points symbolically
x = sy.symbols('x')
solutions = sy.solve(4*x**2 - 7, x)
print(solutions)

# Evaluate for values if desired
print([sol.evalf() for sol in solutions])

# Find a root numerically with sympy
root = sy.nsolve(x - sy.cos(x), 1)
print(root)

# Find a root numerically with scipy
f = lambda x: x - np.cos(x)
sol = sp.optimize.root(f, 1.0)
print(sol.x)


# For dx/dt = tanh(x), plot dx/dt vs x
x = np.linspace(-4, 4, 400)
y = np.tanh(x)

plt.figure(figsize=(2,1))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("Tanh[x]")
plt.show()

# For dx/dt = tanh(x) - x/2, plot tanh(x) and x/2 so that they can be compared
plt.figure(figsize=(2,1))
plt.plot(x, np.tanh(x), label="tanh(x)")
plt.plot(x, x/2, label="x/2")
plt.ylim([-1.2,1.2])
plt.xlabel("x")
plt.ylabel("y")
plt.legend(fontsize=8,loc="lower right",frameon=False,bbox_to_anchor=(1.05, 0.0))
# use a small legend font due to space constraints and shift it a little further right.
plt.show()
```

## Create a matching problem
Use $\frac{dx}{dt} = 4x^2-16$ and ask students to match initial conditions to time series (using their phase portrait from question 1)

```python
# Make time series for matching starting from close to 2 (the unstable fixed point)

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the ODE
def f(t, x):
    return 4*x**2-16

# Time interval and initial condition
t_span = (0, 1)
x0 =[1.95]

# Solve the IVP
sol = solve_ivp(f, t_span, x0, dense_output=True)

# Evaluate solution on a fine time grid
t = np.linspace(t_span[0], t_span[1], 400)
x = sol.sol(t)[0]

# Plot a time series
plt.figure(figsize=(2.5,1.5))
plt.plot(t, x, linewidth=2)

ax = plt.gca()
ax.set_xlabel("t", loc="right")
ax.xaxis.set_label_coords(1.12, -0.07)
plt.ylabel("x(t)")
plt.tight_layout()
time_series_path = "C02_timeseries2A-S26.png"
plt.savefig(time_series_path,dpi=300, bbox_inches='tight')
plt.show()


# Plot another (incorrect) time series
plt.figure(figsize=(2.5,1.5))
plt.plot(t, 3.95*np.exp(-8*t)-2, linewidth=2)

ax = plt.gca()
ax.set_xlabel("t", loc="right")
ax.xaxis.set_label_coords(1.12, -0.07)
plt.ylabel("x(t)")
plt.tight_layout()
time_series_path = "C02_timeseries2B-S26.png"
plt.savefig(time_series_path,dpi=300, bbox_inches='tight')
plt.show()
```
